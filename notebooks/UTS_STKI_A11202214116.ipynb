````python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
import re

# 1. Membuat Stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# 2. Mendapatkan Daftar Stopword Bahasa Indonesia
# 'stopwords' adalah kata-kata umum yang akan dibuang
list_stopword = stopwords.words('indonesian')

# Tambahan: Kadang ada kata-kata umum yang tidak ada di NLTK

print("Stemmer dan daftar stopword siap.")
````

````python
def preprocess_text(text):
    # 1. Case Folding: Ubah semua teks menjadi huruf kecil
    text = text.lower()

    # 2. Cleaning: Hapus karakter non-alfabet (angka, tanda baca)
    # re.sub akan mengganti semua yang BUKAN huruf (a-z) dengan spasi
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 3. Tokenizing: Pecah teks menjadi daftar kata (token)
    # Ini akan menghasilkan list seperti ['fakultas', 'teknik', 'dan', 'informatika', ...]
    tokens = nltk.tokenize.word_tokenize(text)

    # 4. Stopword Removal
    cleaned_tokens = []
    for token in tokens:
        if token not in list_stopword:
            cleaned_tokens.append(token)

    # 5. Stemming
    # Ubah setiap kata di 'cleaned_tokens' ke bentuk dasarnya
    stemmed_tokens = []
    for token in cleaned_tokens:
        stemmed_token = stemmer.stem(token)
        stemmed_tokens.append(stemmed_token)

    # Kita gabungkan kembali token-token yang sudah bersih menjadi satu string
    # Ini akan berguna untuk TfidfVectorizer nanti
    return " ".join(stemmed_tokens)

print("Fungsi preprocess_text() berhasil dibuat.")
````

````python
import pandas as pd

def build_incidence_matrix(corpus, doc_names):
    """
    Membangun Incidence Matrix dari korpus bersih.

    Returns:
    pandas.DataFrame: DataFrame di mana
                      Index = term (vocabulary)
                      Kolom = Nama dokumen
                      Nilai = 1 (ada) atau 0 (tidak ada)
    """

    # --- Langkah 1: Bangun Vocabulary ---
    vocabulary = set()
    for doc in corpus:
        vocabulary.update(doc.split())

    # Urutkan vocabulary agar rapi
    sorted_vocabulary = sorted(list(vocabulary))

    # --- Langkah 2: Buat Matriks (sebagai dictionary dulu) ---
    matrix_data = {}

    # Loop per dokumen
    for doc_id, doc_name in enumerate(doc_names):
        terms_in_doc = set(corpus[doc_id].split())

        # Buat satu kolom untuk dokumen ini
        column = []
        for term in sorted_vocabulary:
            if term in terms_in_doc:
                column.append(1) # 1 jika term ada di dokumen
            else:
                column.append(0) # 0 jika tidak ada

        matrix_data[doc_name] = column

    # --- Langkah 3: Ubah jadi DataFrame ---
    # index=sorted_vocabulary -> membuat barisnya adalah kata-kata
    df_incidence = pd.DataFrame(matrix_data, index=sorted_vocabulary)

    return df_incidence

print("--- SOAL 03 = Incidence Matrix ---")
print("Membangun vocabulary dan incidence matrix...")

incidence_matrix = build_incidence_matrix(korpus_bersih, dokumen_names)

print(f"Berhasil dibuat. Total vocabulary (term): {incidence_matrix.shape[0]}, Dokumen: {incidence_matrix.shape[1]}")
print("\n")

# Tampilkan 15 baris pertama dari matriks
print("Incidence Matrix (Contoh 15 term pertama):")
print(incidence_matrix.head(15))
````

````python
def create_inverted_index(corpus):
    """
    Membuat inverted index dari korpus yang sudah diproses.

    Args:
    corpus (list of str): List berisi dokumen yang sudah bersih (string).

    Returns:
    dict: Dictionary yang mewakili inverted index.
          Key: term (str)
          Value: set of doc_ids (int)
    """
    inverted_index = {}

    # Loop melalui setiap dokumen beserta index-nya (0, 1, 2, ...)
    for doc_id, doc in enumerate(corpus):

        # Pecah dokumen (string) menjadi kata-kata (token)
        terms = doc.split()

        for term in terms:
            # Jika term belum ada di index, buatkan set kosong
            if term not in inverted_index:
                inverted_index[term] = set()

            # Tambahkan doc_id ke dalam set untuk term tersebut
            inverted_index[term].add(doc_id)

    return inverted_index

print("Fungsi create_inverted_index() berhasil dibuat.")
````

````python
# Buat index-nya!
inverted_index = create_inverted_index(korpus_bersih)

# Mari tes hasilnya
print("--- HASIL INVERTED INDEX (CONTOH) ---")

# Tes 1: Kata yang harusnya ada di Dokumen 1 (index 0)
print(f"Kata 'informatika' ada di dokumen: {inverted_index.get('informatika', set())}")

# Tes 2: Kata yang harusnya ada di Dokumen 2 (index 1)
print(f"Kata 'mahasiswa' ada di dokumen: {inverted_index.get('mahasiswa', set())}")

# Tes 3: Coba kita cek kata 'universitas' yang harusnya ada di doc 1 & 2 (index 0 & 1)
# (Berdasarkan output Langkah 8 Anda)
print(f"Kata 'universitas' ada di dokumen: {inverted_index.get('universitas', set())}")

# Tes 4: Tes kata yang tidak ada
print(f"Kata 'olahraga' ada di dokumen: {inverted_index.get('olahraga', set())}")

print("\n")
print(f"Total kata unik (term) di index: {len(inverted_index)}")
````

````python
def search_and(term1, term2):
    """
    Mencari dokumen yang mengandung term1 DAN term2.
    """
    # .get(term, set()) akan mengembalikan set kosong jika term tidak ada
    # Ini mencegah error jika kita mencari kata yang tidak ada di index
    set1 = inverted_index.get(term1, set())
    set2 = inverted_index.get(term2, set())

    # & adalah operator 'intersection' (irisan)
    return set1 & set2

def search_or(term1, term2):
    """
    Mencari dokumen yang mengandung term1 ATAU term2.
    """
    set1 = inverted_index.get(term1, set())
    set2 = inverted_index.get(term2, set())

    # | adalah operator 'union' (gabungan)
    return set1 | set2

def search_not(term1, term2):
    """
    Mencari dokumen yang mengandung term1 BUKAN term2.
    """
    set1 = inverted_index.get(term1, set())
    set2 = inverted_index.get(term2, set())

    # - adalah operator 'difference' (selisih)
    return set1 - set2

# Fungsi helper untuk menampilkan hasil dengan rapi
# (Variabel 'korpus_mentah' dan 'dokumen_names' sudah ada dari Fase 2)
def print_results(doc_ids):
    if not doc_ids:
        print("Tidak ada dokumen yang ditemukan.")
        return

    print(f"Ditemukan {len(doc_ids)} dokumen (ID: {doc_ids}):")
    for doc_id in doc_ids:
        print(f"  - [{dokumen_names[doc_id]}]: {korpus_mentah[doc_id][:100]}...") # Tampilkan 100 char pertama

print("Fungsi search_and(), search_or(), search_not(), dan print_results() siap.")
````

````python
# --- TES 1: Query AND (DAN) ---
# cari: "universitas DAN mahasiswa"
print("--- HASIL QUERY: 'universitas DAN mahasiswa' ---")
# 1. Stem query terms
q_term1 = stemmer.stem("universitas") # -> "universitas"
q_term2 = stemmer.stem("mahasiswa")   # -> "mahasiswa"

# 2. Lakukan pencarian
hasil_and = search_and(q_term1, q_term2)

# 3. Tampilkan hasil
# Ekspektasi: 'universitas' ada di {0, 1}, 'mahasiswa' ada di {1}. Irisannya {1}
print_results(hasil_and)
print("\n")


# --- TES 2: Query OR (ATAU) ---
# cari: "fakultas ATAU mahasiswa"
print("--- HASIL QUERY: 'fakultas ATAU mahasiswa' ---")
q_term1 = stemmer.stem("fakultas")   # -> "fakultas"
q_term2 = stemmer.stem("mahasiswa")  # -> "mahasiswa"
hasil_or = search_or(q_term1, q_term2)
# Ekspektasi: 'fakultas' ada di {0}, 'mahasiswa' ada di {1}. Gabungannya {0, 1}
print_results(hasil_or)
print("\n")


# --- TES 3: Query NOT (BUKAN) ---
# cari: "universitas BUKAN mahasiswa"
print("--- HASIL QUERY: 'universitas BUKAN mahasiswa' ---")
q_term1 = stemmer.stem("universitas") # -> "universitas"
q_term2 = stemmer.stem("mahasiswa")   # -> "mahasiswa"
hasil_not = search_not(q_term1, q_term2)
# Ekspektasi: 'universitas' ada di {0, 1}, 'mahasiswa' ada di {1}. Selisihnya {0}
print_results(hasil_not)
````

````python
# Ini adalah "Kunci Jawaban" yang kita buat secara manual
# Key: Query (sebagai tuple agar bisa jadi key dictionary)
# Value: Set dari doc_id yang kita anggap relevan
ground_truth = {
    ("universitas", "DAN", "mahasiswa"): {1}, # Hanya Doc 2 (index 1) yg relevan
    ("fakultas", "ATAU", "mahasiswa"): {0, 1}, # Doc 1 (index 0) dan Doc 2 (index 1) relevan
    ("universitas", "BUKAN", "mahasiswa"): {0}  # Hanya Doc 1 (index 0) yg relevan
}

print("Ground Truth (Kunci Jawaban) untuk 3 query berhasil dibuat.")
print(ground_truth)
````

````python
def calculate_precision_recall_f1(retrieved_docs, relevant_docs):
    """
    Menghitung Precision, Recall, dan F1-Score.

    Args:
    retrieved_docs (set): Set ID dokumen yang dikembalikan sistem.
    relevant_docs (set): Set ID dokumen yang relevan (dari ground truth).

    Returns:
    tuple: (precision, recall, f1)
    """
    # True Positives (TP): Dokumen yang relevan DAN ditemukan
    # (irisan dari kedua set)
    tp = len(retrieved_docs & relevant_docs)

    # False Positives (FP): Dokumen yang ditemukan TAPI tidak relevan
    # (selisih: yang ada di retrieved tapi tidak ada di relevant)
    fp = len(retrieved_docs - relevant_docs)

    # False Negatives (FN): Dokumen yang relevan TAPI tidak ditemukan
    # (selisih: yang ada di relevant tapi tidak ada di retrieved)
    fn = len(relevant_docs - retrieved_docs)

    # Menghitung Precision
    # Precision = TP / (TP + FP)
    if (tp + fp) == 0:
        precision = 0.0 # Menghindari pembagian dengan nol
    else:
        precision = tp / (tp + fp)

    # Menghitung Recall
    # Recall = TP / (TP + FN)
    if (tp + fn) == 0:
        recall = 0.0 # Menghindari pembagian dengan nol
    else:
        recall = tp / (tp + fn)

    # Menghitung F1-Score
    # F1 = 2 * (Precision * Recall) / (Precision + Recall)
    if (precision + recall) == 0:
        f1 = 0.0 # Menghindari pembagian dengan nol
    else:
        f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

print("Fungsi calculate_precision_recall_f1() berhasil dibuat.")
````

````python
# List untuk menyimpan hasil
all_results = []
total_precision = 0
total_recall = 0
total_f1 = 0

print("--- EVALUASI WAJIB: MODEL BOOLEAN ---")

# Loop melalui setiap query di 'ground_truth'
for query_tuple, relevant_docs in ground_truth.items():
    term1, op, term2 = query_tuple

    # 1. Stem query terms (sama seperti di Langkah 12)
    q_term1 = stemmer.stem(term1)
    q_term2 = stemmer.stem(term2)

    # 2. Jalankan fungsi pencarian yang sesuai
    retrieved_docs = set()
    if op == "DAN":
        retrieved_docs = search_and(q_term1, q_term2)
    elif op == "ATAU":
        retrieved_docs = search_or(q_term1, q_term2)
    elif op == "BUKAN":
        retrieved_docs = search_not(q_term1, q_term2)

    # 3. Hitung metriknya
    precision, recall, f1 = calculate_precision_recall_f1(retrieved_docs, relevant_docs)

    # Simpan total
    total_precision += precision
    total_recall += recall
    total_f1 += f1

    # 4. Tampilkan hasil
    print(f"\nQuery: '{term1} {op} {term2}'")
    print(f"  - Dokumen Ditemukan (Sistem): {retrieved_docs}")
    print(f"  - Dokumen Relevan (Ground Truth): {relevant_docs}")
    print(f"  - Precision: {precision:.2f}")
    print(f"  - Recall: {recall:.2f}")
    print(f"  - F1-Score: {f1:.2f}")

# Hitung Rata-rata
num_queries = len(ground_truth)
avg_precision = total_precision / num_queries
avg_recall = total_recall / num_queries
avg_f1 = total_f1 / num_queries

print("\n--- RATA-RATA METRIK (Macro-Average) ---")
print(f"Average Precision: {avg_precision:.2f}")
print(f"Average Recall: {avg_recall:.2f}")
print(f"Average F1-Score: {avg_f1:.2f}")
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Inisialisasi Vectorizer
# TfidfVectorizer akan otomatis melakukan:
# - Tokenizing (tapi kita sudah, jadi kita berikan string)
# - Menghitung TF
# - Menghitung DF dan IDF
vectorizer = TfidfVectorizer()

# 2. Membuat Matriks TF-IDF
tfidf_matrix = vectorizer.fit_transform(korpus_bersih)

print("--- Matriks TF-IDF (Vector Space Model) ---")
print(f"Bentuk matriks: {tfidf_matrix.shape}")
print("(Artinya: 5 dokumen, 28 kata unik)") # Angka 28 bisa berbeda-beda

print("\nMatriks (dalam format sparse):")
print(tfidf_matrix)
````

````python
# 1. Tentukan query mentah
query_mentah = "informasi mahasiswa"

# 2. Proses query mentah menggunakan fungsi dari Fase 2
query_bersih = preprocess_text(query_mentah)

print(f"Query Mentah: '{query_mentah}'")
print(f"Query Bersih (setelah stem, dll): '{query_bersih}'")

# 3. Ubah query bersih menjadi Vektor TF-IDF
query_vector = vectorizer.transform([query_bersih])

print("\n--- Representasi Vektor Query ---")
print(f"Bentuk vektor query: {query_vector.shape}")
print("(Artinya: 1 query, 51 kolom/kata unik, sama seperti matriks dokumen)")
print(query_vector)
````

````python
from sklearn.metrics.pairwise import cosine_similarity

# 1. Hitung cosine similarity
# Ini membandingkan 1 vektor query dengan 5 vektor dokumen
scores = cosine_similarity(query_vector, tfidf_matrix)

# 2. Tampilkan hasilnya
print("--- Skor Cosine Similarity ---")
print("Bentuk matriks skor:", scores.shape) # Outputnya (1, 5)
print("Skor mentah:", scores)
print("\n")

skor_dokumen = scores[0]

# Tampilkan skor untuk setiap dokumen
print("Skor Kemiripan (0.0 - 1.0) untuk setiap dokumen:")
for i, score in enumerate(skor_dokumen):
    print(f"  - Dokumen {i+1} ({dokumen_names[i]}): {score:.4f}")
````

````python
# Ambil skor dari hasil Langkah 20
# (scores[0] adalah list berisi 5 skor: [0.0, 0.3740, ...])
skor_dokumen = scores[0]

# Tentukan K (jumlah hasil teratas yang ingin ditampilkan)
k = 3

# --- Bagian Inti untuk Ranking ---
# 1. Buat list dari (index_dokumen, skor)
# pakai enumerate(skor_dokumen) untuk mendapatkan (0, 0.0), (1, 0.3740), ...
paired_scores = list(enumerate(skor_dokumen))

# 2. Urutkan list tersebut berdasarkan skor (elemen ke-1 dari tuple)
# reverse=True artinya mengurutkan dari skor TERBESAR ke terkecil
sorted_scores = sorted(paired_scores, key=lambda x: x[1], reverse=True)

# 3. Ambil top-k hasil
top_k_scores = sorted_scores[:k]

# --- Menampilkan Hasil Sesuai Format Soal ---
print(f"--- HASIL TOP-{k} UNTUK QUERY: '{query_mentah}' ---")
print("----------------------------------------------------------------------")
print(f"{'Rank':<5} | {'Doc ID':<12} | {'Score':<10} | {'Snippet (120 char)':<120}")
print("----------------------------------------------------------------------")

# Loop melalui top-k hasil
for rank, (doc_id, score) in enumerate(top_k_scores):
    # Ambil nama file asli dan teks mentah
    nama_file = dokumen_names[doc_id]
    teks_mentah = korpus_mentah[doc_id]

    # Buat snippet (120 karakter pertama)
    snippet = teks_mentah.replace("\n", " ")[:120] + "..."

    # Cetak dalam format tabel
    print(f"{rank+1:<5} | {nama_file:<12} | {score:<10.4f} | {snippet:<120}")
````

````python
# Kunci Jawaban (Ground Truth) untuk Model VSM
# Key: Query (string)
# Value: Set dari doc_id (index) yang relevan

ground_truth_vsm = {
    # Query 1: (Yang baru di uji)
    # Doc 2, 3, 4, 5 semua mengandung "mahasiswa" atau "informasi".
    "informasi mahasiswa": {1, 2, 3, 4},

    # Query 2:
    # Hanya Doc 1 yang mengandung "fakultas" DAN "informatika".
    "fakultas informatika": {0},

    # Query 3:
    # Hanya Doc 3 yang membahas "beasiswa".
    "informasi beasiswa": {2}
}

print("Ground Truth untuk VSM (3 query) berhasil dibuat.")
print(ground_truth_vsm)
````

````python
def calculate_precision_at_k(retrieved_doc_ids, relevant_doc_ids, k):
    """
    Menghitung Precision@k.

    Args:
    retrieved_doc_ids (list): List ID dokumen (terurut) hasil ranking sistem.
    relevant_doc_ids (set): Set ID dokumen yang relevan (ground truth).
    k (int): Batas ranking.
    """
    # Ambil top-k hasil saja
    top_k_retrieved = retrieved_doc_ids[:k]

    # Hitung berapa banyak yang relevan di top-k
    relevant_and_retrieved_at_k = [doc_id for doc_id in top_k_retrieved if doc_id in relevant_doc_ids]

    # P@k = (Jumlah relevan di top-k) / k
    precision_at_k = len(relevant_and_retrieved_at_k) / k
    return precision_at_k

def calculate_average_precision_at_k(retrieved_doc_ids, relevant_doc_ids, k):
    """
    Menghitung Average Precision (AP@k).

    Args:
    retrieved_doc_ids (list): List ID dokumen (terurut) hasil ranking sistem.
    relevant_doc_ids (set): Set ID dokumen yang relevan (ground truth).
    k (int): Batas ranking.
    """
    # Ambil top-k hasil saja
    top_k_retrieved = retrieved_doc_ids[:k]

    # Jika tidak ada yang relevan di ground truth, AP = 0
    if not relevant_doc_ids:
        return 0.0

    score = 0.0
    num_relevant_docs_found = 0

    # Loop per rank (1, 2, 3, ... k)
    for i, doc_id in enumerate(top_k_retrieved):
        rank = i + 1

        # Jika dokumen di rank ini relevan
        if doc_id in relevant_doc_ids:
            num_relevant_docs_found += 1
            # Hitung P@rank saat itu
            precision_at_current_rank = num_relevant_docs_found / rank
            # Tambahkan ke total skor
            score += precision_at_current_rank

    # Jika tidak ada dokumen relevan yang ditemukan, AP = 0
    if num_relevant_docs_found == 0:
        return 0.0

    # Bagi total skor dengan jumlah total dokumen relevan
    # (bisa len(relevant_doc_ids) atau min(len(relevant_doc_ids), k) tergantung definisi)
    # pakai len(relevant_doc_ids) untuk definisi MAP standar
    average_precision = score / len(relevant_doc_ids)

    # Batasi AP tidak lebih dari 1.0 (jika k < jumlah relevan)
    return min(average_precision, 1.0)

print("Fungsi calculate_precision_at_k() dan calculate_average_precision_at_k() siap.")
````

````python
# Tentukan k
k = 3

total_ap_at_k = 0
total_p_at_k = 0

print(f"--- EVALUASI WAJIB: VSM (k={k}) ---")

# Loop melalui 3 query di ground truth
for query_mentah, relevant_docs in ground_truth_vsm.items():

    # --- 1. Jalankan Sistem VSM (sama seperti Langkah 19-21) ---
    query_bersih = preprocess_text(query_mentah)
    query_vector = vectorizer.transform([query_bersih])
    scores = cosine_similarity(query_vector, tfidf_matrix)

    # --- 2. Dapatkan Ranking (ID Dokumen terurut) ---
    # .argsort() mengurutkan, [-1::-1] membalik urutannya (terbesar dulu)
    # .flatten() mengubah [[0.1, 0.3]] menjadi [0.1, 0.3]
    ranked_doc_indices = scores.flatten().argsort()[-1::-1]

    # 'ranked_doc_indices' adalah list terurut, misal: [1, 2, 4, 0, 3]

    # --- 3. Hitung Metrik ---
    p_at_k = calculate_precision_at_k(ranked_doc_indices, relevant_docs, k)
    ap_at_k = calculate_average_precision_at_k(ranked_doc_indices, relevant_docs, k)

    total_p_at_k += p_at_k
    total_ap_at_k += ap_at_k

    # --- 4. Tampilkan Hasil per Query ---
    print(f"\nQuery: '{query_mentah}'")
    print(f"  - Dok. Relevan (GT): {relevant_docs}")
    print(f"  - Dok. Ditemukan (Top-{k}): {ranked_doc_indices[:k]}")
    print(f"  - Precision@{k}: {p_at_k:.4f}")
    print(f"  - Average Precision@{k}: {ap_at_k:.4f}")

# --- 5. Hitung Metrik Rata-rata (MAP@k) ---
num_queries = len(ground_truth_vsm)
map_at_k = total_ap_at_k / num_queries
avg_p_at_k = total_p_at_k / num_queries

print("\n--- RATA-RATA METRIK (Macro-Average) ---")
print(f"Average Precision@{k} (P@{k}): {avg_p_at_k:.4f}")
print(f"Mean Average Precision@{k} (MAP@{k}): {map_at_k:.4f}")
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Inisialisasi Vectorizer BARU dengan Sublinear TF
# Perhatikan parameter sublinear_tf=True
vectorizer_sublinear = TfidfVectorizer(sublinear_tf=True)

# 2. Buat Matriks TF-IDF baru
# etap pakai `korpus_bersih` yang sama
tfidf_matrix_sublinear = vectorizer_sublinear.fit_transform(korpus_bersih)

print("--- VSM dengan Sublinear TF Scaling ---")
print(f"Bentuk matriks: {tfidf_matrix_sublinear.shape}")
print("Vectorizers baru ('vectorizer_sublinear') berhasil dibuat.")
````

````python
# Tentukan k (sama dengan sebelumnya)
k = 3

total_ap_at_k_sublinear = 0
total_p_at_k_sublinear = 0

print(f"--- EVALUASI WAJIB: VSM (Skema Sublinear TF, k={k}) ---")

# Loop melalui 3 query di ground truth VSM kita
for query_mentah, relevant_docs in ground_truth_vsm.items():

    # --- 1. Jalankan Sistem VSM (Versi Sublinear) ---
    query_bersih = preprocess_text(query_mentah)

    # Gunakan vectorizer & matriks BARU
    query_vector = vectorizer_sublinear.transform([query_bersih])
    scores = cosine_similarity(query_vector, tfidf_matrix_sublinear)

    # --- 2. Dapatkan Ranking (ID Dokumen terurut) ---
    ranked_doc_indices = scores.flatten().argsort()[-1::-1]

    # --- 3. Hitung Metrik ---
    # Fungsi evaluasi kita (dari eval.py) bisa dipakai ulang
    p_at_k = calculate_precision_at_k(ranked_doc_indices, relevant_docs, k)
    ap_at_k = calculate_average_precision_at_k(ranked_doc_indices, relevant_docs, k)

    total_p_at_k_sublinear += p_at_k
    total_ap_at_k_sublinear += ap_at_k

    # --- 4. Tampilkan Hasil per Query ---
    print(f"\nQuery: '{query_mentah}'")
    print(f"  - Dok. Relevan (GT): {relevant_docs}")
    print(f"  - Dok. Ditemukan (Top-{k}): {ranked_doc_indices[:k]}")
    print(f"  - Precision@{k}: {p_at_k:.4f}")
    print(f"  - Average Precision@{k}: {ap_at_k:.4f}")

# --- 5. Hitung Metrik Rata-rata (MAP@k) ---
num_queries = len(ground_truth_vsm)
map_at_k_sublinear = total_ap_at_k_sublinear / num_queries
avg_p_at_k_sublinear = total_p_at_k_sublinear / num_queries

print("\n--- RATA-RATA METRIK (Skema Sublinear TF) ---")
print(f"Average Precision@{k} (P@{k}): {avg_p_at_k_sublinear:.4f}")
print(f"Mean Average Precision@{k} (MAP@{k}): {map_at_k_sublinear:.4f}")
````
